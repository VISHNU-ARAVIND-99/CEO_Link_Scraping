```python
from bing_results_scraper import BingScraper
import pandas as pd

def linkedin_scraping(company_name):
  links = [company_name]
  bing = BingScraper()  # headers=Optional(default header will be used), max_retries=Optional(default: 3), num_results=Optional(default: 10)
  if "University" not in company_name and "College" not in company_name:
    results = bing.get_results(q=f"{company_name} + Executive Team + Leadership Team + Chief Executive Officer")
  else:
    results = bing.get_results(q=f'{company_name} + Executive Team + Leadership Team + President + CEO')
  if results is not None and 'organic_results' in results:
    for i in results["organic_results"]:
      if i["link"]:
        links.append(i["link"])
  return links


list_of_list = []
with open('company.txt', 'r') as file:
    links = file.readlines()
name_list = [x.strip() for x in links]


out = 0
for i in name_list:
    list_of_list.append(linkedin_scraping(company_name=i))
    out += 1
    print(out)
    df = pd.DataFrame(list_of_list)
    df.to_excel('Company_Executive_Page_Link_Scraper_Out.xlsx', index=False)
```
